```python
#imports
import math
import numpy as np
import pandas as pd
import os
import sys
import subprocess
import matplotlib as plt
import seaborn as sns
%matplotlib qt
```

```python
#dataframes to load
trunc_xls=['phase1_waterfall_trun.xlsx','phase2_waterfall_trun.xlsx','phase2_waterfall_trun.xlsx']
full_xls=['phase1_waterfall_full.xlsx','phase2_waterfall_full.xlsx','phase3_waterfall_full.xlsx']
    

trunc_dfs={}
for xls in trunc_xls:
    trunc_dfs[xls]=pd.read_excel(xls)
```

```python
#load in csvs instead of excel files
#excel writer doesnt write full dataframe because reasons

phase1full=pd.read_csv('phase1_waterfall_full.csv')
phase2full=pd.read_csv('phase2_waterfall_full.csv')
phase3full=pd.read_csv('phase3_waterfall_full.csv')

phase1full.index=phase1full.loc[:,'Unnamed: 0']
phase2full.index=phase2full.loc[:,'Unnamed: 0']
phase3full.index=phase3full.loc[:,'Unnamed: 0']

phase1full = phase1full.drop('Unnamed: 0',axis=1)
phase2full = phase2full.drop('Unnamed: 0',axis=1)
phase3full = phase3full.drop('Unnamed: 0',axis=1)

#set up dict
full_dfs={}
full_dfs['phase1full']=phase1full
full_dfs['phase2full']=phase2full
full_dfs['phase3full']=phase3full

#shift/fix frequencies so there is no unwanted alphabetical sorting
#shift phase1 freqs by 0.8 * 0.015625
shifted_freaks=[]
for i in phase1full.columns:
    shifted_freaks.append(np.float(i)+(0.8*0.015625))
phase1full.columns=shifted_freaks

#change phase 1 and 2 columns to floats
float_f2=[]
for i in phase2full.columns:
    float_f2.append(np.float(i))
phase2full.columns=float_f2

float_f3=[]
for i in phase3full.columns:
    float_f3.append(np.float(i))
phase3full.columns=float_f3

#entire data of all 3 phases
all_phases=pd.concat(full_dfs,axis=0)
#all_phases.index=range(0,83)
```
